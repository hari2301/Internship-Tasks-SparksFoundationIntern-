***Importing packages***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

***Reading the data***

```python
data=pd.read_csv('SampleSuperstore.csv')
data.head(4)
```

***Describing and visualizing***

```python
data.describe().T
```

```python
data.shape
```

```python
data.isnull().sum()
```

#### profit based on category

Technology category provides more profit

```python
plt.bar(data['Category'],data['Profit'])
```

```python
#unique values for all columns
data.nunique()
```

#### States counts

States the rapidly consumes the products

```python
plt.figure(figsize=[20,10])
sns.countplot(x=data['State'])
plt.xticks(rotation=90)
plt.show();
```

#### Sub-catogaries count

sub categories that consumes more

```python
plt.figure(figsize=[20,10])
sns.countplot(x=data['Sub-Category'])
plt.xticks(rotation=90)
plt.show();
```

#### Discount and profit variation

```python
plt.figure(figsize=[15,7])
sns.lineplot('Discount','Profit',data=data,color='r')
```

#### Sale by categories

```python
plt.bar(data['Category'],data['Sales'])
```

#### Profit produced by sub-categories: 

```python
plt.figure(figsize=[20,9])
plt.bar(data['Sub-Category'],data['Profit'])
```

#### Sales based on region

```python
plt.bar(data['Region'],data['Sales'])
```

#### profit based on region

```python
plt.bar(data['Region'],data['Profit'])
```

#### Profit based on ship mode

```python
plt.bar(data['Ship Mode'],data['Profit'])
```

#### Sales and profit 

```python
plt.figure(figsize=[10,6])
sns.scatterplot(data['Sales'],data['Profit'],hue='Profit',data=data)
```
